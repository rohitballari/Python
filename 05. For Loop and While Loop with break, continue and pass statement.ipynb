### What is Loop in Python?

In Python, a loop is a control structure that allows you to execute a block of code repeatedly. Python supports two types of loops: the **while loop** and the **for loop**. Both loops allow you to iterate over a block of code multiple times, but they are used in different scenarios.

### While Loop:



#### The while loop in Python is used to execute a block of code repeatedly as long as a given condition is true. It is a fundamental control structure that allows for repetitive tasks and can be useful in various scenarios, such as iterating over data until a certain condition is met.

### Let's start with a simple example of a while loop that prints numbers from 1 to 5:

###    Steps Involved in While Loop
- Initialization: The variable count is initialized to 1.
- Condition: The loop checks the condition count <= 5. If this condition is True, the loop executes the indented block of code.
- Execution: The loop prints the current value of count and then increments count by 1 using count += 1.
- Re-evaluation: The condition is re-evaluated. If count is still less than or equal to 5, the loop continues.
- Termination: Once count becomes greater than 5, the condition count <= 5 becomes False, and the loop terminates.

```python
a=10
while a<20:
    print(a)
    a=a+1
    #a+=1
```

```python
count = 1                  # Initialization: The variable count is initialized to 1.
while count <= 5:          #Condition: The loop checks the condition count <= 5. If this condition is True, the loop executes the indented block of code.
    print(count)
    count += 1             # Execution: The loop prints the current value of count and then increments count by 1 using count += 1
```

### Breaking Out of a Loop

```python
count = 1
while count <= 5:
    if count == 3: 
        break      
    print(count)
    count += 1
```

```python
a = 10
while a <= 100:
    if a >= 15:
        break
    print('I Love Machine Learning' ,a)
    a+=1
```

### Continue of a While Loop

The continue statement skips the rest of the code inside the loop for the current iteration and jumps back to the beginning of the loop for the next iteration. 


```python
count = 0
while count < 5:
    count += 1
    if count == 3:
        continue
    print(count)
```

### Nested while Loops


```python
outer_count = 1
while outer_count <= 3:
    inner_count = 1
    while inner_count <= 2:
        print(f"Outer: {outer_count}, Inner: {inner_count}")
        inner_count += 1
    outer_count += 1
```

### Using else with while loop
Python allows you to use an else block with a while loop. The else block is executed when the loop condition becomes False

```python
count = 1
while count <= 3:
    print(count)
    count += 1
else:
    print("Loop is finished")
```

### The range() Function

The range() function returns a sequence of numbers, starting from 0 by default, and increments by 1 (by default), and ends at a specified number.

```python
for i in range(6):    # Note that range(6) is not the values of 0 to 6, but the values 0 to 5.
    print(i)
```

The range() function defaults to increment the sequence by 1, however it is possible to specify the increment value by adding a third parameter: range(start,stop,step):  
-->By default the step size =1

### Nested Loops
Nested loop is a loop inside a loop          
The inner loop will be executed one time for each iteration of the outer loop

```python
name = ['Jeevan','Jagruti','Neha','Chandana']
subjects= ['Mysql','Python','Stats']
print('Name , Subjects')
for i in name:
    for j in subjects:
        print(i,'--->',j)
    print('i am going for outer loop')
```
